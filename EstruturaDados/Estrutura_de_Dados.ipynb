# Listas


#### Manipulação de Listas

```python
cidades = ['Rio de Janeiro', 'São Paulo', 'Salvador', 'Goiania']
#                0               1            2            3

print(cidades[2])
print(cidades[-2])

cidades [0] = 'Brasilia'

print(cidades[0])
```

#### Funções dentro de Listas

```python
cidades = ['Rio de Janeiro', 'São Paulo', 'Salvador', 'Goiania']

print('Adiciona')
cidades.append('Curitiba')
print(cidades)

print('Remove')
cidades.remove('Rio de Janeiro')
print(cidades)

print('Insere na posição')
cidades.insert(1, 'Rio de Janeiro')
print(cidades)

print('Remove a posição')
cidades.pop(0)
print(cidades)

print('Organizar em ordem alfabética')
cidades.sort()
print(cidades)
```

#### Concatenando Listas

```python
num = [2, 3, 4, 5]
letra = ['a', 'b', 'c']

# final = num + letra
# print(final)
num.extend(letra)
print(num)
```

#### Unpacking Lists

```python
produtos = ['arroz', 'feijao', 'banana', 'laranja']

item1, item2, item3, *outros = produtos

print(produtos[0])
```

#### Looping em Listas

```python
valor = [10, 20, 30, 40, 50]

for num in valor:
  print(f'O valor final do produto é de {num}.')
```

#### Verificando itens em uma lista

```python
cliente = str(input('Digite a cor desejada: '))
cores = ['Amarelo', 'Verde', 'Azul', 'Vermelho']

if cliente in cores:
  print('Em estoque.')
else:
  print('Sem estoque.')
```

#### Agregando duas listas com Zip

```python
cores = ['Amarelo', 'Verde', 'Azul', 'Vermelho']
valor = [10, 20, 30, 40]

duas_listas = zip(cores, valor)
print(list(duas_listas))
```

# Sets
- Similar a listas

- Evita itens duplicados

- Não utiliza index


```python
list1 = [10, 20, 30, 40, 50]
list2 = [10, 20, 60, 70 ]

num1 = set(list1)
num2 = set(list2)

print(num1 | num2) # União
print(num1 - num2) # Diferença
print(num1 ^ num2) # Diferença Simétrica
print(num1 & num2) # And
print(len(num1)) # Tamanho
```

#### Set com Números


```python
s1 = {1, 2, 3, 4, 5, 6, 1, 2}

s1.add(7)
s1.update([6, 7, 8, 9, 10])
s1.remove(10)
s1.discard(9)

print(s1)
```

#### Set com Strings

```python
set1 = {'a', 'b', 'c'}
set2 = {'a', 'd', 'e'}
set3 = {'c', 'd', 'f'}

set4 = set1.union(set2)
set5 = set1.difference(set3)
set6 = set1.intersection(set2)

print(set4, set5, set6)
```

# Função Lambda

#### É uma função (pequena) e sem nome, pode ter vários argumentos, mas somente uma expressão. Muito utilizada dentro de outras funções, clean code.

```python
# Função normal, com nome.
def soma(x):
  return x + 10
soma(2)

# Função Lambda
somar10 = lambda x,y: x + y + 10
print(somar10(2, 4))
```

#### Lambda dentro de outra função

```python
def somar(x):
  func2 = lambda x: x + 10
  return func2(x) * 4

print(somar(2))
```

# Função Map
#### Muito utilizado com listas, aplicar uma função a uma Iterable, por item. (list, tuple, dic, etc.)

```python
lista1 = [1, 2, 3, 4]

def mult(x):
  return x * 2

lista2 = map(mult, lista1)
print(list(lista2))
```

#### Map com Lambda

```python
lista1 = [1, 2, 3, 4]

print(list( map(lambda x: x * 2, lista1)))
```

# Filter
#### Muito utilizado com listas, aplicar uma função a um Iterable, filtrando os items. (list, tuple, dic, etc.)

```python
valores = [10, 12, 34, 44, 57]

# def remover20(x):
#   return x > 20
# print(list(map(remover20, valores)))
print(list(filter(lambda x: x > 20, valores)))
```

# Generator Expressions
#### Uma forma mais rápida para listas, dicionários, etc. Menos memória alocada, valores em bytes.

```python
from sys import getsizeof

numeros = [x * 10 for x in range(1000)]
print(type(numeros))
print(numeros)
print(getsizeof(numeros))

print('======')

numeros = (x * 10 for x in range(1000))
print(type(numeros))
print(list(numeros))
print(getsizeof(numeros))
```
