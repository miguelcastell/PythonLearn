## Classe


```python
class Casa: # Classe
  def __init__(self, cor, quarto, banheiro): # Construtor
    self.cor = cor
    self.quarto = quarto
    self.banheiro = banheiro


  def mostrar_cor(self): # Métodos
    print(f'A cor da casa é {self.cor}')

  def mostrar_quarto(self):
    print(f'Esta casa tem {self.quarto}')

  def mostrar_banheiro(self):
    print(f'Possue {self.banheiro} banheiro(s)')

  def adicionar_quarto(self):
    self.quarto += 1
    print(f'Agora essa casa tem {self.quarto} quartos.')

  def pintar_casa(self, nova_cor):
    print(f'Pintando a casa de {self.cor} para {nova_cor}')

casa1 = Casa('Azul', 5, 2) # Instâncias
casa2 = Casa('Amarela', 3, 1)

print('\nCasa 1: ')
casa1.mostrar_cor()
casa1.mostrar_quarto()
casa1.mostrar_banheiro()
casa1.adicionar_quarto()
casa1.pintar_casa('Verde')

print('\nCasa 2: ')
casa2.mostrar_cor()
casa2.mostrar_quarto()
casa2.mostrar_banheiro()
casa2.adicionar_quarto()
casa2.pintar_casa('Vermelho')
```

Criando uma classe da Carros no Collab

```python
class Carro:
  def __init__(self, cor, ano, marca, direcao):
    self.cor = cor
    self.ano = ano
    self.marca = marca
    self.ligado = False
    self.seta = direcao

  def informacoes(self):
    print('Segue as informações sobre o carro desejado.')
    print(f'A cor do carro é: {self.cor}.')
    print(f'O ano do carro é: {self.ano}.')
    print(f'A marca é: {self.marca}.')

  def ligar(self):
    if not self.ligado:
      self.ligado = True
      print('O carro foi ligado')
    else:
      print('O carro já estava ligado.')

  def desligar(self):
    if self.ligado:
      self.ligado = False
      print('O carro foi desligado.')
    else:
      print('O carro já estava desligado.')

  def ligar_seta(self, direcao):
    if not self.ligado:
      print('Ligue o carro primeiro.')
      return

    self.seta = direcao
    print(f'A seta foi ligado {self.seta}')

carro1 = Carro('Preto', '2017', 'BMW', 'Centro')
carro2 = Carro('Azul', '2026', 'Honda', 'Centro')

print('Carro 1:')
carro1.informacoes()
carro1.ligar()
carro1.desligar()
carro1.ligar()
carro1.ligar_seta('Esquerda')

print('\nCarro 2:')
carro2.informacoes()
carro2.ligar()
carro2.desligar()
carro2.ligar_seta('Direita')
```

```python
class Pessoa:
  def __init__(self, nome, idade, cargo):
    self.nome = nome
    self.idade = idade
    self.cargo = cargo

  def info (self):
    print(f'O nome da pessoa é {self.nome}.')
    print(f'A idade é de {self.idade} anos.')
    print(f'O seu cargo atual é {self.cargo}.')

  def promover (self, novo_cargo):
    print(f'{self.nome} foi promovido e agora exerce a função de {novo_cargo}.')
    self.cargo = novo_cargo


pessoa1 = Pessoa('Miguel', '18', 'Analista de Sistemas')
pessoa2 = Pessoa('Taina', 19, 'Estiagiária em T.I.')

print('Pessoa 1:')
pessoa1.info()
pessoa1.promover('Diretor de Sistemas')

print('\nPessoa 2:')
pessoa2.info()
pessoa2.promover('Analista de Dados Sênior')
```

## Herança

```python
class Animal:
  def __init__(self, nome, cor, especie):
    self.nome = nome
    self.cor = cor
    self.especie = especie

  def apresentar(self):
    print(f'Eu sou o {self.especie} chamado {self.nome}, minha cor é {self.cor}.')

class Gato (Animal):
    def emitir_Som(self):
      print(f'O gato está fazendo miauuu!')

class Cachorro (Animal):
    def emitir_Som(self):
      print('O cachorro está fazendo AuAu!')

class Elefante (Animal):
    def emitir_Som(self):
      print('O elefante ta fazendo Pruuuuu!')

gato1 = Gato('Felix', 'Preto', 'Siamês')
gato1.apresentar()
gato1.emitir_Som()

cachorro1 = Cachorro('Amora', 'Branca com manchas marrom', 'Shit-Zu')
cachorro1.apresentar()
cachorro1.emitir_Som()

elefante1 = Elefante('Dumbo', 'Cinza escuro', 'Orelhudo')
elefante1.apresentar()
elefante1.emitir_Som()
```

```python
class Pessoa:
  def __init__(self, nome, idade, local):
    self.nome = nome
    self.idade = idade
    self.local = local

  def apresenta (self):
    print(f'Olá! Me chamo {self.nome}, tenho {self.idade} anos e sou de {self.local}.')

class Funcionario (Pessoa):
  def __init__(self, nome, idade, local, funcao):
    super().__init__(nome, idade, local)
    self.funcao = funcao

  def trabalho(self):
    print(f'Eu trabalho como {self.funcao}')

class Cliente (Pessoa):
  pass

funcionario1 = Funcionario('Miguel', '18', 'Itapira - SP', 'Analista')
funcionario1.apresenta()
funcionario1.trabalho()

cliente1 = Cliente('Ana', 25, 'São Paulo')
cliente1.apresenta()
```

## Herança Múltipla


```python
# Classe Avo
class Animal():
  def __init__(self, nome):
    self.nome = nome

# Classes Pai
class Predador(Animal):
  def cacando(self):
    print(f'O Animal {self.nome} está caçando.')

class Presa(Animal):
  def fugindo(self):
    print(f'O Animal {self.nome} está sendo caçado.')

# Classes Filho
class Coelho(Presa):
  pass

class Tigre(Predador):
  pass

class Golfinho(Predador, Presa):
  pass

coelho1 = Coelho('Bunny')
tigre1 = Tigre('Tiger')
golfinho1 = Golfinho('Dolphin')

coelho1.fugindo()
tigre1.cacando()
golfinho1.cacando()
```

# Função Super()


```python
class Escola():
  def __init__(self, nome, idade, status):
    self.nome = nome
    self.idade = idade
    self.status = status

  def Apresentar(self):
    print(f'Meu nome é {self.nome}.')

  def verificar_status(self):
    print(f'Status: {'ATIVO' if self.status else "INATIVO"}')

class Aluno(Escola):
  def __init__(self, nome, idade, status, ano):
    super().__init__(nome, idade, status)
    self.ano = ano

  def Apresentar(self):
    super().Apresentar()
    print('Eu sou um Aluno da escola.')

class Professor(Escola):
  def __init__(self, nome, idade, status, materia):
    super().__init__(nome, idade, status)
    self.materia = materia

  def Apresentar(self):
    super().Apresentar()
    print('Eu sou um Professor da escola.')

class Assistente(Escola):
  def __init__(self, nome, idade, status, bloco):
    super().__init__(nome, idade, status)
    self.bloco = bloco
  def Apresentar(self):
    super().Apresentar()
    print('Eu sou um Assistente da escola.')

a1 = Aluno(nome='Miguel', idade=12, status=True, ano=8)
p1= Professor('Júlio', 31, False, 'HardCode')
as1 = Assistente('Márcia', 25, True, 'Bloco C')

a1.verificar_status()
```

## Polymorphism



```python
class Personagens():
  def falar(self):
    print('Eu sou um personagem.')

class Guerreiro(Personagens):
  def falar(self):
    print('Eu sou um guerreiro forte e destemido!')

class Mago(Personagens):
  def falar(self):
    print('Eu sou um mago sábio e poderoso!')

class Arqueiro(Personagens):
  def falar(self):
    print('Eu sou um arqueiro rápido e preciso!')

# Criar os objetos:
personagens = [Guerreiro(), Mago(), Arqueiro()]

for p in personagens:
  p.falar()
```

## Agregação


```python
class Motor:
  def __init__(self, marca, potencia):
    self.marca = marca
    self.potencia = potencia

class Carro:
  def __init__(self):
    self.motores = []

  def adicionar_motor(self, motor):
    self.motores.append(motor)

  def listar_motores(self):
    for motor in self.motores:
      print(f'Marca: {motor.marca} - {motor.potencia} cavalos de potencia')

# Criando os motores (objetos).
motor_v6 = Motor('Ford', 300)
motor_v8 = Motor('Ferrari', 650)
motor_v12 = Motor('Lamborghini', 950)

# Criar o carro e add o motor nele.
carro1 = Carro()
carro1.adicionar_motor(motor_v6)
carro1.adicionar_motor(motor_v8)
carro1.adicionar_motor(motor_v12)

# Listar os motores
carro1.listar_motores()
```

```python
#Exercício

class Livro:
  def __init__(self, autor, categoria, status):
    self.autor = autor
    self.categoria = categoria
    self.status = status

  def verificar_status(self):
    print('Status:','ATIVO' if self.status else 'INATIVO')

class Biblioteca:
  def __init__(self):
    self.livros = []

  def add_livro(self, livro):
    self.livros.append(livro)

  def li_livro(self):
    for livro in self.livros:
      print(f'O autor do livro é o(a) {livro.autor} e está categorizado como {livro.categoria}.')

# Criando os livros
up = Livro(autor='Disney', categoria='Aventura', status=True)
starwars = Livro(autor='Disney', categoria='Ficção', status=False)

# Criando a Biblioteca
deparis = Biblioteca()
deparis.add_livro(up)

daitalia = Biblioteca()
daitalia.add_livro(starwars)

# Listando Livro
deparis.li_livro()
up.verificar_status()

daitalia.li_livro()
starwars.verificar_status()
```

## Nasted Classes


```python
class Computador:
  def __init__(self, modelo, gpu_nome, gpu_memoria, cpu_nome, cpu_cores, cpu_clock, ram_nome, ram_gb):
    self.modelo = modelo
    self.gpu= self.GPU(gpu_nome, gpu_memoria)
    self.cpu = self.CPU(cpu_nome, cpu_cores, cpu_clock)
    self.ram = self.RAM(ram_nome, ram_gb)


  def mostrar_configuracao(self):
    print(f'Computador: {self.modelo}.')
    self.gpu.mostrar_gpu()
    self.cpu.mostrar_cpu()
    self.ram.mostrar_ram()

  class GPU: # Nasted Class
    def __init__(self, nome, memoria_gb):
      self.nome = nome
      self.memoria_gb = memoria_gb

    def mostrar_gpu(self):
      print(f'GPU: {self.nome} - {self.memoria_gb}GB.')

  class CPU:
    def __init__(self, nome, cores, clock_ghz):
      self.nome = nome
      self.cores = cores
      self.clock_ghz = clock_ghz

    def mostrar_cpu(self):
      print(f'CPU: {self.nome} - {self.cores} núcleos - {self.clock_ghz}GHz')

  class RAM:
    def __init__(self, nome, gb):
      self.nome = nome
      self.gb = gb

    def mostrar_ram(self):
      print(f'RAM: {self.nome} - {self.gb}GB')

# Utilização
pc1 = Computador('Dell XPS', 'NVIDIA RTX 4090', 24, 'CPU Intel Core i7', 8, 4.6, 'XPG Gammix', 16)
pc1.mostrar_configuracao()
```
