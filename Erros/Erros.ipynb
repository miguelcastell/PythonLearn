# Erros
#### Excelente para testes, não realiza o 'stop' no programa, mensagens customizados quando encontra um erro.

#### Try e Except com lista.

```python
try:
  letras = ['a', 'b', 'c']
  print(letras[3])
except IndexError:
  print('Index não existe.')
```

#### Try e Except com input.

```python
try:
  valor = int(input('Digite o valor do produto: '))
  print(valor)
except ValueError:
  print('Digite um número inteiro.')
```

#### Else e Finally

```python
try:
  valor = int(input('Digite o valor do produto: '))
  print(valor)
except ValueError:
  print('Digite um número inteiro.')
finally:
  print('Codigo ok.')

# else:
#   print('Usuário digitou um valor correto.')
#   resultado = valor * 2
#   print(resultado)
```
