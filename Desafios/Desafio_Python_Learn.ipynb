```python
# Desafio do Steak

temp = float(input('Digite a temperatura da carne em Graus Celsius: '))

if temp <= 48:
  print('A carne está selada ou crua.')
elif 48 < temp <= 54:
  print('A carne está ao ponto para mal.')
elif 54 < temp <= 60:
  print('A carne está ao ponto.')
elif 60 < temp <= 65:
  print('A carne está ao ponto para bem.')
else:
  print('A carne está bem passada.')
```

```python
# Calculadora para Pintura
import math

rendimento = float(input('Digite o rendimento em m²: '))
altura = float(input('Digite a altura da parede em m: '))
largura = float(input('Digite a largura da parede em m: '))

def calculo_tinta():
  tinta = (altura * largura) / rendimento
  lata = math.ceil(tinta)
  print(f'Você necessita de de {lata:.0f} latas de tinta pra printar a parede.')

calculo_tinta()
```

```python
# Filtrando Lista

funcionarios = [' Ana', 'Marcos', 'Alice', 'Pedro', 'Sophia', 'Bruno', 'Melissa']
turno_dia = ['Ana', 'Marcos', 'Alice', 'Melissa']
turno_noite = ['Pedro', 'Sophia', 'Bruno']
tem_carro = ['Marcos', 'Alice', 'Bruno',' Melissa']

lista1 =  set(tem_carro).intersection(turno_noite)
print(lista1)

lista2 =  set(tem_carro).intersection(turno_dia)
print(lista2)

lista3 =  set(funcionarios).difference(tem_carro)
print(lista3)
```

```python
# Cálculo IMC

altura = float(input('Digite sua altura: '))
peso = float(input('Digite seu peso: '))
imc = peso / (altura * altura)

if imc <= 18.5:
  print('Magreza')
elif 18.5 < imc <= 24.9:
  print('Normal')
elif 25.0 < imc <= 29.9:
  print('Sobrepeso')
elif 30.0 < imc <= 39.9:
  print('Obesidade')
else:
  print('Obesidade grave')
```

```python
# Desafio 26

def potencia(base, expoente=2):
  return base ** expoente

user_number = int(input(' Digite o numero base: '))
user_expoente = input('Digite um expoente (default 2):')

if user_expoente:
  print(f'O resultado é: {potencia(user_number, int(user_expoente))}')
else:
  print(f'O resultado é: {potencia(user_number)}')
```

```python
# Desafio 27

def fatorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n * fatorial(n - 1)

user_number = int(input('Digite um número: '))
print(f'O fatorial de {user_number} é {fatorial(user_number)}')
```
