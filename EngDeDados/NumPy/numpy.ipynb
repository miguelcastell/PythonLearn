# Array

### Dimensionamento de Arrays

```python
# Array 1D

d1 = np.array([1,2,3])
print(d1.shape)
print(d1.ndim)
```

```python
# Array 2D

d2 = np.array([[1,2,3], [4,5,6]])
print(d2.shape)
print(d2.ndim)
```

```python
# Array 3D

d3 = np.array([[[1,2], [3,4]], [[5,6], [7,8]]])
print(d3.shape)
print(d3.ndim)
```
