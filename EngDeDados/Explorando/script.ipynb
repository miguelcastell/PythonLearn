# Filtrando Dados

```python
filter1 = df.loc[df['new_hires'] == 0]
filter1
```

```python
filter2 = df.loc[df['new_hires'] > 2000]
filter2
```

```python
filter3 = df.loc[(df['new_hires'] > 2000) & (df['layoffs'] == 1070)]
filter3
```

```python
import re
filter4 = df.loc[df['company'].str.contains('AMD| Samsung', flags=re.I)]
filter4
```

```python
filter5 = df.loc[
    (df['year'] >= 2000) &
    (df['year'] < 2010) &
    (df['company'] == 'AMD')
]

filter5
```

# Modificando Dados


```python
df.loc[df['confidence_level'] == 'Medium', 'confidence_level'] = 'Médio'
filter5
```
