## Descrição das Colunas do Arquivo `data1.csv`

Este documento detalha as colunas presentes no conjunto de dados `data1.csv`.

| Coluna | Descrição |
|---|---|
| `company` | O nome da empresa à qual os dados se referem. |
| `year` | O ano em que os dados foram registrados. |
| `employees_start` | O número de funcionários no início do ano. |
| `employees_end` | O número de funcionários no final do ano. |
| `new_hires` | O número de novos funcionários contratados durante o ano. |
| `layoffs` | O número de funcionários demitidos durante o ano. |
| `net_change` | A variação líquida no número de funcionários (contratações - demissões). |
| `hiring_rate_pct` | A taxa de contratação como uma porcentagem do número inicial de funcionários. |
| `attrition_rate_pct` | A taxa de atrito (desgaste) de funcionários como uma porcentagem. |
| `revenue_billions_usd` | A receita da empresa em bilhões de dólares americanos. |
| `stock_price_change_pct` | A variação percentual do preço das ações da empresa. |
| `gdp_growth_us_pct` | A taxa de crescimento do Produto Interno Bruto (PIB) dos Estados Unidos em porcentagem. |
| `unemployment_rate_us_pct` | A taxa de desemprego nos Estados Unidos em porcentagem. |
| `is_estimated` | Um valor booleano que indica se os dados da linha são uma estimativa (`True`) ou não (`False`). |
| `confidence_level` | O nível de confiança na precisão dos dados (por exemplo, Alto, Médio, Baixo). |
| `data_quality_score` | Uma pontuação que representa a qualidade geral dos dados na linha. |
| `employee_growth_pct` | A taxa de crescimento do número de funcionários em porcentagem. |
| `layoff_ratio` | A proporção de demissões em relação ao número de funcionários. |
