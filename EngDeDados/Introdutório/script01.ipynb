```python
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_csv('../data/data.csv')
df.head()
```

# Comandos de Visualização.

```python
print('----Ver colunas.----')
display(df.columns)
print('----Ver quantidade de Indexs.----')
display(df.index)
print('----Ver colunas específicas.----')
display(df[['company', 'year']])
print('----Localizar----')
display(df.iloc[1:5])
```

```python
print('----Filtrar Colunas x Dado----')
display(df.loc[df['company'] == 'AMD'])
print('----Sortear employees_end decrescente----')
display(df.sort_values('employees_end', ascending=False))
```
