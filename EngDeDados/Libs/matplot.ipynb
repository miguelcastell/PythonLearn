# Matplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/gas_prices.csv')
df.head
```

```python
plt.title('Valores do Combustível em US$')

# plt.plot(df['Year'], df['Australia'], 'r.--',label = 'Australia')
# plt.plot(df['Year'], df['Italy'], 'y.--', label = 'Italy')
# plt.plot(df['Year'], df['USA'], 'b.--', label = 'USA')
# plt.plot(df['Year'], df['Japan'], 'g.--',label = 'Japan')
list_country = ['Australia', 'USA', 'Canada']

for country in df:
    if country in list_country:
        plt.plot(df['Year'], df[country], label=[country], marker='.')
    

plt.xticks(df['Year'][::2])

plt.xlabel('Years')
plt.ylabel('US$')

plt.legend()
plt.show()

plt.savefig('data/country_gas.png')
```
